## 0. An end-to-end Scikit-Learn  workflow

```python
# 1. Get the data ready
```

```python
# Standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
heart_disease = pd.read_csv('heart_disease-Copy1.csv')
heart_disease
```

```python
# X is features matrix
X = heart_disease.drop("target", axis=1)

# y is labels
y = heart_disease["target"]
```

```python
# 2. Choose the right model 
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We will keep default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train);
```

```python
# make a prediction
y_label = clf.predict(np.array([1,2,3,4]))
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve the model
# try different amount of n_estimators

np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
```

```python
# 6. Save a model and load it
import pickle

pickle.dump(clf, open("random_forest_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pk1", "rb"))
loaded_model.score(X_test, y_test)
```

#### Fill missing data with pandas

```python
# Fill the "Make" columns
car_sales_missing_data["Make"].fillna("missing", inplace=True)

# Fill the "Colour" column
car_sales_missing_data["Colour"].fillna("missing", inplace=True)

# Fill the "Odometer (KM)" column
car_sales_missing_data["Odometer (KM)"].fillna(car_sales_missing_data["Odometer (KM)"].mean(), inplace=True)

# Fill the "Doors"
car_sales_missing_data["Doors"].fillna(4, inplace=True)
```
